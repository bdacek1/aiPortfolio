```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
import pandas as pd
```

```python
df = pd.read_csv('archive/Musical_Instruments_reviews.csv')
df.head()
```

```python
nltk.tokenize.word_tokenize(df['summary'][0])
```

```python
stemmer = nltk.stem.PorterStemmer()
stemmed = [stemmer.stem(word) for word in nltk.tokenize.word_tokenize(df['summary'][55])]
print(stemmed)
```

```python
lemmatizer = nltk.stem.WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in nltk.tokenize.word_tokenize(df['summary'][55])]
print(lemmatized)
```

```python
rows = []
stemmer = nltk.stem.PorterStemmer()
lemmatizer = nltk.stem.WordNetLemmatizer()
for i in range(len(df['summary'])):
    summary = df.loc[i]['summary']
    token = nltk.tokenize.word_tokenize(summary)
    stemmed = [stemmer.stem(word) for word in nltk.tokenize.word_tokenize(summary)]
    lemmatized = [lemmatizer.lemmatize(word) for word in nltk.tokenize.word_tokenize(summary)]
    row = [summary, token, stemmed, lemmatized]
    rows.append(row)
```

```python
new_df = pd.DataFrame(rows, columns = ["Original", "Tokenized", "Stemmed", "Lemmatized"])
```
