```python
import pandas as pd
from sklearn.feature_extraction import FeatureHasher
```

```python
df = pd.read_csv('cars.csv', low_memory=False) #Read in the Data
df.head()
```

```python
for col in df.columns: #Get number of unique values in each column
    print(col, df[col].nunique())
```

Of the features that deal with categorical data, there are two that have far more unique values than the rest;
manufacturer_name and model_name. The rest of the columns we can either one hot if there are more than two unique
values or simply replace the values with 0 or 1 if there are only 2 unique values. For the manufacturer and model name columns, we will want to use the FeatureHasher on them in order to extract information

```python
#These columns all only have two unique values so we can replace their values with 0 and 1
binary_cols = ['engine_has_gas', 'transmission', 'has_warranty', 'is_exchangeable']
for i in range(10):
    binary_cols.append(f'feature_{i}')
```

```python
sparse_cols = ['color','engine_fuel','engine_type','body_type','state','drivetrain','location_region']
```

```python
d = pd.get_dummies(df, columns=sparse_cols) #One hot encode the columns with sparse unique values
d.head()
```

```python
for b in binary_cols: #Replace values in binary columns with 0s and 1s
    unique = d[b].unique()
    d[b] = d[b].replace([unique[0],unique[1]], [0, 1])
```

Using the FeatureHasher since the model_name and manufacturer_name columns have too many unique values to one hot.
Choosing 15 features for the model_name as we don't want too many columns for it otherwise it will be weighed far more
heavily than the other features. 15 keeps it comparable to other columns while still being larger as to prevent too many
collisions. 10 was chosen for the manufacturer_name as it has fewer unique than the model_name.

```python
model_hasher = FeatureHasher(n_features=15, input_type='string')
model_t = model_hasher.transform(d['model_name'])
model_df = pd.DataFrame(model_t.toarray(), columns=[f"md_{i}" for i in range(15)])
n_df = pd.concat([d,model_df], axis=1)
```

```python
manu_hasher = FeatureHasher(n_features=10, input_type='string')
manu_t = manu_hasher.transform(d['manufacturer_name'])
manu_df = pd.DataFrame(manu_t.toarray(), columns=[f"mn_{i}" for i in range(10)])
new_df = pd.concat([n_df,manu_df], axis=1)
```

```python
new_df.drop('manufacturer_name', inplace=True,axis=1)
new_df.drop('model_name', inplace=True,axis=1)
```

```python
new_df.to_csv('transformed_cars.csv')
```
